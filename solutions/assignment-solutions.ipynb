### Question 1:

Write a Python function that takes a PySpark DataFrame `df` and a column name `col` as input, and returns a new DataFrame with a new column `new_col` that contains the square root of the values in `col`.


**Answer:**

```python
# answer code
from pyspark.sql.functions import sqrt

def add_sqrt_column(df, col):
    new_df = df.withColumn("new_col", sqrt(col))
    return new_df
```

```python
new_df = add_sqrt_column(df, "a")
new_df.show()
```

### Question 3

Write a PySpark code snippet to filter a DataFrame `df` to only include rows where the value in the `column1` column is greater than 10 and the value in the `column2` column contains the substring "foo".

```python
# Create Sample Data
data = [
    (5, "foobar"), 
    (15, "hello foo world"), 
    (8, "foodie"), 
    (20, "foo bar"), 
    (25, "not_a_match"), 
    (30, "foo123")
]

# Define Schema
columns = ["column1", "column2"]

# Create DataFrame
df = spark.createDataFrame(data, columns)
```

```python
from pyspark.sql.functions import col, instr

filtered_df = df.filter((col("column1") > 10) & (instr(col("column2"), "foo") > 0))
filtered_df.show()
```
